## <a id='toc1_'></a>1. Introduction [&#8593;](#toc0_)

**Project Overview**

This project analyzes personality, lifestyle, and demographic data to help a dating app understand which attributes of a user pair indicate romantic compatibility. The dataset includes 100,000 pair records with Big Five personality traits, career information, lifestyle preferences, communication style, love language, demographics, **gender, and sexual orientation** for both individuals, along with a target column called `compatible` which indicates whether the pair is compatible.

**Objective (Ideal Partner Range)**

Use the dataset to train a classification model on the `compatible` target, then apply that model in reverse: given a **single user's profile**, score them against real partner profiles from a **20% holdout partner pool** (reserved before model training) and identify the **ranges of partner traits** that most consistently lead to compatibility. The output is an "ideal partner profile" with recommended ranges, ranked by importance.

**Workflow**

The workflow progresses from EDA, through preprocessing and feature engineering, to modeling, threshold tuning, and feature importance. **Section 8** takes a single user's input, scores them against orientation-filtered profiles from the holdout partner pool using the trained model, and extracts the compatible partner ranges.

**Dataset Update**

This notebook uses the **modified dataset** (`cupid_algorithm_dataset_sexualorient_added.csv`) which adds 4 new columns: `a_gender`, `a_sexual_orientation`, `b_gender`, `b_sexual_orientation`. Gender and sexual orientation are used as **pre-filters** (to ensure logically valid pairings) and as **categorical features** in the model. See `Risks_Adding_Sexual_Orientation_Feature.md` for risk assessment.

**Data Dictionary for Data file**

    Per-person features (x2 for a_ and b_):
    - age: individual's age (18-55)
    - openness: Big Five personality trait - openness to experience (0-1)
    - extraversion: Big Five personality trait - extraversion (0-1)
    - agreeableness: Big Five personality trait - agreeableness (0-1)
    - conscientiousness: Big Five personality trait - conscientiousness (0-1)
    - career_field: individual's career field (10 industries)
    - career_ambition: individual's ambition level (0-1)
    - chronotype: sleep preference, night owl to morning person (0-1)
    - spontaneity: planning preference, planner to free spirit (0-1)
    - emotional_expressiveness: emotional expressiveness level (0-1)
    - love_language: preferred love language (5 categories)
    - education: education level (1-5)
    - location: location type (Urban, Suburban, Rural)
    - gender: individual's gender (Male, Female) [NEW]
    - sexual_orientation: individual's sexual orientation (Straight, Gay, Lesbian, Bisexual) [NEW]
    
    Target variables:
    - compatible: Binary target indicating whether the pair is compatible (0/1)
    - compatibility_score: Continuous compatibility score (0-100)
    - relationship_longevity_months: Predicted relationship duration (0-120 months)


**Evaluation Framework**

Because there is class imbalance in the `compatible` column (43% compatible / 57% not compatible), model performance will be evaluated using the F1 score to balance precision and recall on the minority class (ie `compatible`). Additional analysis will include inspection of the confusion matrix and ROC curve, with particular focus on how threshold changes alter the sensitivity of the matching system and the trade-off between false positives and false negatives.

### <a id='toc4_5_'></a>4_5 DCP: Feature Engineering - Encoding categorical features[&#8593;](#toc0_)

```python
#Identifying categorical columns that need one-hot encoding
ohe_cols = [col for col in cupid_df.columns if cupid_df[col].dtype == 'object']
print(f"Categorical columns to one-hot encode: {ohe_cols}")
print(f"\nUnique values per categorical column:")
for col in ohe_cols:
    print(f"  {col}: {cupid_df[col].nunique()} unique -> {cupid_df[col].unique()}")
```

```python
# Instantiate the OneHotEncoder
one_hot_encoder = OneHotEncoder(sparse_output=False, handle_unknown="ignore")

#Fit and tranform Categorical columns for one hot encoding
one_hot_encoder.fit(cupid_df[ohe_cols])
ary_encoded_catg_features = one_hot_encoder.transform(cupid_df[ohe_cols])

#Get encoded feature names
encoded_feature_names = one_hot_encoder.get_feature_names_out()


#Creating DataFrame from encoded data
df_encoded_catg_features = pd.DataFrame(ary_encoded_catg_features, columns = encoded_feature_names, index=cupid_df.index)

#Replacing original cat columns with one hot encoded cat columns
cupid_ohe_df = pd.concat([cupid_df.drop(columns = ohe_cols), df_encoded_catg_features], axis =1)

#inspecting complete df with one hot encoded features
display(cupid_ohe_df.sample(5))

#Checking size
print(f"\n The size of the cupid_df is {cupid_df.shape} \n") 

#Checking size
print(f"\n The size of the cupid_ohe_df is {cupid_ohe_df.shape} \n") 

#Checking all feature names in encoded df to verify inclusion of encoded feature name
display(cupid_ohe_df.columns)
```

### <a id='toc4_6_'></a>4_6 DCP: Feature Engineering - Creating delta and match features[&#8593;](#toc0_)

```python
# Creating delta features - how similar/different are the pair's traits?
# These measure the absolute difference between person a and person b for each numeric trait
# '_delta' or '_gap' are for difference between a_ and b_ Numerical features , while '_same' are for the categorical variables

# Personality deltas
for trait in ['openness', 'extraversion', 'agreeableness', 'conscientiousness']:
    cupid_ohe_df[f'{trait}_delta'] = abs(cupid_ohe_df[f'a_{trait}'] - cupid_ohe_df[f'b_{trait}'])

# Lifestyle deltas
cupid_ohe_df['chronotype_delta'] = abs(cupid_ohe_df['a_chronotype'] - cupid_ohe_df['b_chronotype'])
cupid_ohe_df['spontaneity_delta'] = abs(cupid_ohe_df['a_spontaneity'] - cupid_ohe_df['b_spontaneity'])
cupid_ohe_df['ambition_delta'] = abs(cupid_ohe_df['a_career_ambition'] - cupid_ohe_df['b_career_ambition'])
cupid_ohe_df['expressiveness_delta'] = abs(cupid_ohe_df['a_emotional_expressiveness'] - cupid_ohe_df['b_emotional_expressiveness'])

# Age gap
cupid_ohe_df['age_gap'] = abs(cupid_ohe_df['a_age'] - cupid_ohe_df['b_age'])

# Creating match flags - do they share the same categorical trait?
cupid_ohe_df['same_career'] = (cupid_df['a_career_field'] == cupid_df['b_career_field']).astype(int)
cupid_ohe_df['same_love_language'] = (cupid_df['a_love_language'] == cupid_df['b_love_language']).astype(int)
cupid_ohe_df['same_location'] = (cupid_df['a_location'] == cupid_df['b_location']).astype(int)
cupid_ohe_df['same_education'] = (cupid_df['a_education'] == cupid_df['b_education']).astype(int)
cupid_ohe_df['same_gender'] = (cupid_df['a_gender'] == cupid_df['b_gender']).astype(int)
cupid_ohe_df['same_orientation'] = (cupid_df['a_sexual_orientation'] == cupid_df['b_sexual_orientation']).astype(int)

# Composite similarity scores
cupid_ohe_df['personality_similarity'] = cupid_ohe_df[['openness_delta', 'extraversion_delta', 'agreeableness_delta', 'conscientiousness_delta']].mean(axis=1)
cupid_ohe_df['lifestyle_similarity'] = cupid_ohe_df[['chronotype_delta', 'spontaneity_delta']].mean(axis=1)

#Viewing the new features
delta_cols = [col for col in cupid_ohe_df.columns if 'delta' in col or 'same_' in col or 'similarity' in col or 'age_gap' in col]
print(f"\nNew engineered features ({len(delta_cols)}): {delta_cols}")
display(cupid_ohe_df[delta_cols].describe())

#Checking size
print(f"\n The size of the cupid_ohe_df after feature engineering is {cupid_ohe_df.shape} \n")
```

## <a id='toc5_'></a> 5. EDA [&#8593;](#toc0_)

This EDA addresses three questions designed to identify the strongest compatibility predictors, test whether they compound, and understand what compensates when the top predictor is unfavorable.


### <a id='toc5_1_'></a>5_1 EDA Q1: What are the strongest predictors of compatibility, and how do they rank? [&#8593;](#toc0_)

We compute compatibility rates by match/mismatch for every categorical feature, and correlation with compatibility for every trait comparison feature, then produce a single ranked table of all factors by correlation.


```python
# Q1 Part A: Categorical feature compatibility rates (matched vs mismatched)
# For each match flag, compute the compatibility rate when matched vs not matched

catg_var_interest = ['same_love_language', 'same_career', 'same_location',
                     'same_education', 'same_gender', 'same_orientation']

catg_results = []

for col in catg_var_interest:
    matched_rate = cupid_ohe_df[cupid_ohe_df[col] == 1]['compatible'].mean() * 100
    mismatched_rate = cupid_ohe_df[cupid_ohe_df[col] == 0]['compatible'].mean() * 100
    gap = matched_rate - mismatched_rate

    catg_results.append({
        'Trait comparison feature (Categorical)': col,
        'Matched Rate (%)': round(matched_rate, 1),
        'Mismatched Rate (%)': round(mismatched_rate, 1),
        'Gap (%)': round(gap, 1)
    })

catg_df = pd.DataFrame(catg_results).sort_values('Gap (%)', ascending=False)

print("Categorical feature compatibility rates (matched vs mismatched):")
print("=" * 75)
display(catg_df)

print(f"\nThere is also a higher compatibility rate ({catg_df.iloc[0]['Matched Rate (%)']:.1f} vs {catg_df.iloc[0]['Mismatched Rate (%)']:.1f}) for {catg_df.iloc[0]['Trait comparison feature (Categorical)']}. That's a {catg_df.iloc[0]['Gap (%)']:.1f} difference in compatibility rate")
```

```python
# Q1 Part B: Combined ranking of ALL Trait comparison features by Correlation
# Trait comparison features ranked by absolute correlation

# Compute correlation between each delta feature and the compatible target
num_var_interest = delta_cols + ['compatible']
corr_with_target = cupid_ohe_df[num_var_interest].corr()['compatible'].drop('compatible').sort_values()

# keep only numerical Trait comparison feature names, ie those that end with '_delta' or '_gap'
delta_corr = corr_with_target[corr_with_target.index.str.endswith('_delta', 'gap')]

# keep only categorical Trait comparison feature names, ie those that end with '_same'
same_corr = corr_with_target[corr_with_target.index.str.startswith('same_')]

combined_ranking = []

# Add trait comparison features
for feat, corr_val in corr_with_target.items():
    combined_ranking.append({
        'Trait comparison feature': feat,
        'Correlation': f"r = {corr_val:.3f}",
        'Abs Effect': abs(corr_val)
    })

ranking_df = pd.DataFrame(combined_ranking).sort_values('Abs Effect', ascending=False)

print("Combined Predictor Ranking (all features, sorted by Correlation):")
print("=" * 90)
display(ranking_df[['Trait comparison feature', 'Correlation']].reset_index(drop=True))

# Identify the top categorical and top trait comparison feature predictor for Q2
top_categorical = same_corr.abs().idxmax()
top_numerical = delta_corr.abs().idxmax()
top_traitcomparison_feat = corr_with_target.abs().idxmax()  # largest absolute correlation
top_traitcomparison_corr = corr_with_target[top_traitcomparison_feat]

print(f"\nTop categorical trait comparison predictor: {top_categorical}")
print("")
print(f"\nTop numerical trait comparison predictor: {top_numerical}")
print("")
print(f"\nOverall Top trait comparison predictor:  {top_traitcomparison_feat} (r = {top_traitcomparison_corr:.3f})")
```

```python
# Q1 Visualization: Compatibility rates for categorical features (matched vs mismatched)

fig = go.Figure()

for _, row in catg_df.iterrows():
    feat = row['Trait comparison feature (Categorical)']
    fig.add_trace(go.Bar(name='Matched', x=[feat], y=[row['Matched Rate (%)']],
                         marker_color='steelblue', showlegend=(_ == 0)))
    fig.add_trace(go.Bar(name='Mismatched', x=[feat], y=[row['Mismatched Rate (%)']],
                         marker_color='salmon', showlegend=(_ == 0)))

fig.update_layout(
    title='Compatibility Rate: Matched vs Mismatched Categorical Features',
    yaxis_title='Compatibility Rate (%)',
    barmode='group',
    height=450,
    width=900
)
fig.show()

# Correlation bar chart for trait comparison features
fig2 = go.Figure()
fig2.add_trace(go.Bar(
    y=[f for f in corr_with_target.index],
    x=[v for v in corr_with_target.values],
    orientation='h',
    marker_color=['salmon' if v < 0 else 'steelblue' for v in corr_with_target.values]
))
fig2.update_layout(
    title='Correlation of Trait Comparison Features with Compatibility',
    xaxis_title='Correlation (r)',
    height=450,
    width=900
)
fig2.show()
```

```python
# -------------------------------------------------------
# Finding: EDA Q1 - Strongest Predictors of Compatibility
# -------------------------------------------------------

print("**-------------------------------------------------------**")
print()
print("Finding: EDA Q1 - Strongest Predictors of Compatibility")
print()
print(f"The single largest compatibility factor is {top_categorical}.")
print(f"  Matched: {catg_df.iloc[0]['Matched Rate (%)']:.1f}% vs Mismatched: {catg_df.iloc[0]['Mismatched Rate (%)']:.1f}%")
print(f"  Gap: {catg_df.iloc[0]['Gap (%)']:.1f} percentage points")
print()
print(f"Among trait comparison features, {top_traitcomparison_feat} has the strongest")
print(f"correlation with compatibility (r = {top_traitcomparison_corr:.3f}).")
print()

print("**-------------------------------------------------------**")
```

### <a id='toc5_2_'></a>5_2 EDA Q2: Do the top predictors compound when both are aligned? [&#8593;](#toc0_)

Using the top categorical and top trait comparison predictor from Q1, we isolate pairs where **both** are aligned. We compare that group's compatibility rate against each predictor aligned individually to determine whether the combined effect is **additive** (sum of individual gains) or **diminishing** (less than the sum).


```python
# Q2: Compounding effect of top two predictors

# Define alignment for the top numerical trait comparison predictor using median split
top_num_median = cupid_ohe_df[top_traitcomparison_feat].median()

# Create alignment flags
cupid_ohe_df['_top_catg_aligned'] = cupid_ohe_df[top_categorical] == 1
cupid_ohe_df['_top_num_aligned'] = cupid_ohe_df[top_numerical] <= top_num_median  # lower delta = more aligned

# Baseline: overall compatibility rate
baseline_rate = cupid_ohe_df['compatible'].mean() * 100

# Individual alignment rates
catg_only_rate = cupid_ohe_df[cupid_ohe_df['_top_catg_aligned']]['compatible'].mean() * 100
num_only_rate = cupid_ohe_df[cupid_ohe_df['_top_num_aligned']]['compatible'].mean() * 100

# Both aligned
both_aligned = cupid_ohe_df[cupid_ohe_df['_top_catg_aligned'] & cupid_ohe_df['_top_num_aligned']]
both_rate = both_aligned['compatible'].mean() * 100

# Individual gains over baseline
catg_gain = catg_only_rate - baseline_rate
num_gain = num_only_rate - baseline_rate
sum_of_gains = catg_gain + num_gain
both_gain = both_rate - baseline_rate

# Additive vs diminishing
effect_type = "additive" if both_gain >= sum_of_gains * 0.95 else "diminishing"


print(f"Compounding Analysis: {top_categorical} + {top_numerical}")
print("=" * 65)
print(f"  Baseline (all pairs):                {baseline_rate:.1f}%")  #Baseline is the average of compatibility rates for each aligned Trait Comparison Features
print(f"  {top_categorical} aligned only:       {catg_only_rate:.1f}% (baseline gain: +{catg_gain:.1f}%)")
print(f"  {top_numerical} aligned only:  {num_only_rate:.1f}% (baseline gain: +{num_gain:.1f}%)")
print(f"  Both aligned:                         {both_rate:.1f}% (baseline gain: +{both_gain:.1f}%)")
print(f"")
print(f"  Sum of individual baseline gains:              +{sum_of_gains:.1f}%")
print(f"  Actual combined baseline gain:                 +{both_gain:.1f}%")
print(f"  Effect type:                          {effect_type}")
print(f"")
print(f"  Both-aligned group size: {len(both_aligned):,} pairs ({len(both_aligned)/len(cupid_ohe_df)*100:.1f}% of dataset)")

# Clean up temp columns
cupid_ohe_df.drop(columns=['_top_catg_aligned', '_top_num_aligned'], inplace=True)
```

```python
# Q2 Visualization: Compounding effect bar chart

categories = ['Baseline\n(all pairs)', f'{top_categorical}\naligned only',
              f'{top_numerical}\naligned only', 'Both\naligned']

fig = go.Figure()

# Baseline
fig.add_trace(go.Bar(
    name='Baseline',
    x=[categories[0]],
    y=[baseline_rate],
    marker_color='gray',
    text=[f'{baseline_rate:.1f}%'],
    textposition='outside'
))

# categorical aligned + numerical aligned ( stacked bar)
fig.add_trace(go.Bar(
    name=f'{top_categorical} aligned only',
    x=[categories[1]],
    y=[catg_only_rate],
    marker_color='steelblue',
    text=[f'{catg_only_rate:.1f}%'],
    textposition='outside'
))

fig.add_trace(go.Bar(
    name=f'{top_numerical} aligned only',
    x=[categories[1]],
    y=[num_only_rate],
    marker_color='mediumpurple',
    text=[f'{num_only_rate:.1f}%'],
    textposition='outside'
))

# Both ALigned bar for both aligned
fig.add_trace(go.Bar(
    name='Both aligned',
    x=[categories[2]],
    y=[both_rate],
    marker_color='green',
    text=[f'{both_rate:.1f}%'],
    textposition='outside'
))

# Expected additive line
expected_additive = baseline_rate + sum_of_gains
fig.add_hline(
    y=expected_additive,
    line_dash="dash",
    line_color="orange",
    annotation_text=f"Expected if additive: {expected_additive:.1f}%",
    annotation_position="top right"
)

fig.update_layout(
    barmode='stack',
    title=f'Compounding Effect: {top_categorical} + {top_numerical}',
    yaxis_title='Compatibility Rate (%)',
    height=450,
    width=700,
    yaxis=dict(range=[0, max(both_rate, expected_additive) * 1.15])
)

fig.show()
```

```python
# -------------------------------------------------------
# Finding: EDA Q2 - Compounding Effect
# -------------------------------------------------------

print("**-------------------------------------------------------**")
print()
print("Finding: EDA Q2 - Do the Top Predictors Compound?")
print()
print(f"When both {top_categorical} and {top_numerical} are aligned,")
print(f"the compatibility rate reaches {both_rate:.1f}% — a {both_gain:.1f}% gain")
print(f"over the baseline of {baseline_rate:.1f}%.")
print()
if effect_type == "additive":
    print(f"The combined gain ({both_gain:.1f}%) is approximately equal to the sum")
    print(f"of individual gains ({sum_of_gains:.1f}%), indicating an ADDITIVE effect.")
    print(f"Both signals contribute independently — neither subsumes the other.")
else:
    print(f"The combined gain ({both_gain:.1f}%) is less than the sum of individual")
    print(f"gains ({sum_of_gains:.1f} pp), indicating a DIMINISHING effect.")
    print(f"There is some overlap between what these two signals capture,")
    print(f"but the combined rate is still substantially higher than either alone.")
print()
print("Implication for CupidMatch: Matching on multiple compatibility signals")
print("produces better outcomes than any single factor alone.")
print()
print("**-------------------------------------------------------**")
```

### <a id='toc5_3_'></a>5_3 EDA Q3: When the #1 predictor is unfavorable, which secondary factors best compensate? [&#8593;](#toc0_)

We filter to pairs where the top predictor from Q1 is misaligned. Among that subset, we compare compatible vs. incompatible pairs across all remaining features to identify which secondary signals most differentiate the two groups.


```python
# Q3: Secondary factors that compensate when top predictor is misaligned

# Filter to pairs where the top categorical predictor is misaligned
misaligned_df = cupid_ohe_df[cupid_ohe_df[top_categorical] == 0].copy()

misaligned_compat_rate = misaligned_df['compatible'].mean() * 100
n_misaligned = len(misaligned_df)
n_misaligned_compat = misaligned_df['compatible'].sum()

print(f"Pairs where {top_categorical} is misaligned: {n_misaligned:,}")
print(f"  Compatible: {n_misaligned_compat:,} ({misaligned_compat_rate:.1f}%)")
print(f"  Incompatible: {n_misaligned - n_misaligned_compat:,} ({100 - misaligned_compat_rate:.1f}%)")
print()

# Compare compatible vs incompatible on all remaining trait comparison features
remaining_deltas = [col for col in delta_cols if col != top_categorical and 'similarity' not in col]

rescue_results = []

for feat in remaining_deltas:
    compat_mean = misaligned_df[misaligned_df['compatible'] == 1][feat].mean()
    incompat_mean = misaligned_df[misaligned_df['compatible'] == 0][feat].mean()
    diff = incompat_mean - compat_mean  # positive = compatible pairs have lower delta (more similar)

    rescue_results.append({
        'Feature': feat,
        'Compatible Mean': round(compat_mean, 4),
        'Incompatible Mean': round(incompat_mean, 4),
        'Difference': round(diff, 4),
        'Abs Difference': round(abs(diff), 4)
    })

rescue_df = pd.DataFrame(rescue_results).sort_values('Abs Difference', ascending=False)

print(f"Secondary factors among {top_categorical}-misaligned pairs:")
print(f"(Sorted by how much they differentiate compatible vs incompatible)")
print("=" * 80)
display(rescue_df[['Feature', 'Compatible Mean', 'Incompatible Mean', 'Difference']].reset_index(drop=True))
```

```python
# Q3 Part B: Categorical rescue factors among misaligned pairs

remaining_catg = [col for col in catg_var_interest if col != top_categorical]

catg_rescue_results = []

for col in remaining_catg:
    # Among misaligned pairs, what's the compat rate when this secondary factor IS matched?
    matched_rate = misaligned_df[misaligned_df[col] == 1]['compatible'].mean() * 100
    mismatched_rate = misaligned_df[misaligned_df[col] == 0]['compatible'].mean() * 100
    gap = matched_rate - mismatched_rate

    catg_rescue_results.append({
        'Feature': col,
        'Matched Rate (%)': round(matched_rate, 1),
        'Mismatched Rate (%)': round(mismatched_rate, 1),
        'Rescue Gap (%)': round(gap, 1)
    })

catg_rescue_df = pd.DataFrame(catg_rescue_results).sort_values('Rescue Gap (%)', ascending=False)

print(f"Categorical rescue factors among {top_categorical}-misaligned pairs:")
print("=" * 75)
display(catg_rescue_df.reset_index(drop=True))
```

```python
# Q3 Visualization: Top rescue factors (trait comparison + categorical combined)

# Combine rescue signals into one view
fig = make_subplots(rows=1, cols=2, subplot_titles=[
    'Trait Comparison: Compatible vs Incompatible Mean Delta',
    'Categorical: Rescue Gap (%)'
], horizontal_spacing=0.15)

# Left: trait comparison rescue factors (top 5)
top_rescue = rescue_df.head(5)
fig.add_trace(go.Bar(
    y=top_rescue['Feature'], x=top_rescue['Compatible Mean'],
    name='Compatible', marker_color='steelblue', orientation='h',
    showlegend=True
), row=1, col=1)
fig.add_trace(go.Bar(
    y=top_rescue['Feature'], x=top_rescue['Incompatible Mean'],
    name='Incompatible', marker_color='salmon', orientation='h',
    showlegend=True
), row=1, col=1)

# Right: categorical rescue gaps
fig.add_trace(go.Bar(
    y=catg_rescue_df['Feature'], x=catg_rescue_df['Rescue Gap (%)'],
    marker_color=['green' if v > 0 else 'salmon' for v in catg_rescue_df['Rescue Gap (%)']],
    orientation='h', showlegend=False
), row=1, col=2)

fig.update_layout(
    title=f'Secondary Factors That Rescue Compatibility When {top_categorical} is Misaligned',
    height=400, width=1100,
    barmode='group'
)
fig.show()
```

```python
# -------------------------------------------------------
# Finding: EDA Q3 - Rescue Factors
# -------------------------------------------------------

top_rescue_continuous = rescue_df.iloc[0]
top_rescue_categorical = catg_rescue_df.iloc[0] if len(catg_rescue_df) > 0 else None

print("**-------------------------------------------------------**")
print()
print(f"Finding: EDA Q3 - Rescue Factors When {top_categorical} is Misaligned")
print()
print(f"Among pairs where {top_categorical} is misaligned (N={n_misaligned:,}),")
print(f"the baseline compatibility rate drops to {misaligned_compat_rate:.1f}%.")
print(f"However, {n_misaligned_compat:,} pairs are still compatible. What distinguishes them?")
print()
print(f"Top trait comparison rescue factor: {top_rescue_continuous['Feature']}")
print(f"  Compatible pairs have a mean of {top_rescue_continuous['Compatible Mean']:.4f}")
print(f"  vs incompatible pairs at {top_rescue_continuous['Incompatible Mean']:.4f}")
print(f"  (difference: {top_rescue_continuous['Difference']:.4f})")
print()
if top_rescue_categorical is not None:
    print(f"Top categorical rescue factor: {top_rescue_categorical['Feature']}")
    print(f"  When matched: {top_rescue_categorical['Matched Rate (%)']:.1f}% compatible")
    print(f"  When mismatched: {top_rescue_categorical['Mismatched Rate (%)']:.1f}% compatible")
    print(f"  Rescue gap: {top_rescue_categorical['Rescue Gap (%)']:.1f}%")
    print()
print(f"Insight: Even when the primary compatibility signal, {top_categorical},")
print(f"is absent, secondary factors can partially compensate")
print()
print(f"Once consideration to improve the models performance is to weight these secondary features")
print(f"more heavily for users whose {top_categorical} cannot be matched in the partner pool dataset.")
print()
print("**-------------------------------------------------------**")
```

### <a id='toc6_0_'></a>6_0 Reserving Holdout Partner Pool (80/20 Split) [&#8593;](#toc0_)

Before performing the train/test split for model building, we first reserve **20% of the dataset** as a **holdout partner pool**. This pool will be used in Section 8 (Ideal Partner Range Generator) instead of generating synthetic hypothetical partners.

**Why a holdout pool instead of synthetic partners:**

1. **Realistic feature correlations** — Synthetic partners are generated by sampling each feature independently (e.g., `b_openness`, `b_career_ambition`, `b_chronotype` are each drawn from uniform distributions). This can produce trait combinations that are unlikely in reality (e.g., extremely high conscientiousness paired with extremely low career ambition). The holdout pool preserves the natural co-occurrence patterns present in the data.

2. **More defensible** — Saying "we scored the user against real profiles from our dataset" is stronger than "we generated synthetic partners" — especially since the Cupid's Algorithm dataset is already synthetic. Stacking synthetic-on-synthetic adds a layer that could draw scrutiny in a presentation.

3. **Larger partner pool** — A 20% holdout of 100K pairs yields ~20K pairs, from which we can extract up to ~40K individual profiles (both Person A and Person B sides). This provides a substantial pool of realistic partner candidates.

4. **Simpler implementation** — No random generation logic or orientation-aware pairing rules needed for partner creation. We simply extract existing profiles and filter by orientation.

**Important:** The holdout pool is split *before* the train/test split, ensuring that no partner pool profiles leak into the model's training or test data.


```python
# Reserve 20% of the dataset as a holdout partner pool BEFORE train/test split
# This ensures no data leakage between the partner pool and model training/test data

from sklearn.model_selection import train_test_split

# Split: 80% for modeling (train + validation + test), 20% for partner pool
modeling_df, partner_pool_df = train_test_split(
    cupid_ohe_df,
    test_size=0.20,
    random_state=22,
    stratify=cupid_ohe_df['compatible']
)

print(f"Total dataset:       {len(cupid_ohe_df):,} pairs")
print(f"Modeling set (80%):  {len(modeling_df):,} pairs")
print(f"Partner pool (20%):  {len(partner_pool_df):,} pairs")
print(f"")
print(f"Partner pool target distribution:")
print(partner_pool_df['compatible'].value_counts().to_string())
print(f"")

# Extract individual profiles from the partner pool
# Each pair has a Person A and Person B — we extract both sides as potential partners
a_cols = [col for col in partner_pool_df.columns if col.startswith('a_')]
b_cols = [col for col in partner_pool_df.columns if col.startswith('b_')]

# Extract Person B profiles (keep b_ prefix — these are partner candidates)
partner_profiles_b = partner_pool_df[b_cols].copy()

# Extract Person A profiles and rename a_ to b_ (so all partners share the same column names)
partner_profiles_a = partner_pool_df[a_cols].copy()
partner_profiles_a.columns = [col.replace('a_', 'b_') for col in a_cols]

# Combine both sides into a single partner pool
holdout_partner_pool = pd.concat([partner_profiles_b, partner_profiles_a], ignore_index=True)

# Remove duplicate profiles (same person may appear in multiple pairs)
holdout_partner_pool = holdout_partner_pool.drop_duplicates().reset_index(drop=True)

print(f"Individual partner profiles extracted: {len(holdout_partner_pool):,}")
print(f"  (after deduplication from {len(partner_profiles_b) + len(partner_profiles_a):,} raw profiles)")
print(f"")

print(f"Partner pool columns: {len(holdout_partner_pool.columns)}")
print(f"Sample columns: {holdout_partner_pool.columns[:10].tolist()}")
```

```python
# Seperate X and y variables
# Using the MODELING set (80%) — the partner pool (20%) has been reserved for Section 8
# Dropping the target and secondary target variables from the features
X = modeling_df.drop(['compatible', 'compatibility_score', 'relationship_longevity_months', 'pair_id'],axis=1)
y = modeling_df['compatible']


# Remainder test split
X_rem, X_test, y_rem, y_test = train_test_split(X, y, test_size=0.3, random_state=22, stratify=y)

# Print info on how the data has been split
print(f'Full dataset: {len(cupid_ohe_df):,} pairs')
print(f'  Partner pool (reserved): {len(partner_pool_df):,} pairs (20%)')
print(f'  Modeling set: {len(modeling_df):,} pairs (80%)')
print(f'    Remainder set: {len(X_rem):,} data points')
print(f'    Test set: {len(X_test):,} data points\n')

#Train Validation split
X_train, X_val, y_train, y_val = train_test_split(X_rem, y_rem, test_size=0.2, random_state=22,stratify = y_rem)

#Checking sizes
print(f'The train set has {len(X_train)} data points.')
print(f'The validation set has {len(X_val)} data points.')
```

```python
# Saving the feature column names for use in Section 8 (ideal partner generation)
feature_columns = X.columns.tolist()
print(f"Total features for modeling: {len(feature_columns)}")
```

### <a id='toc7_1_'></a>7_1 Baseline Model: Using a Simple Logistic Regression Model as a Baseline Model[&#8593;](#toc0_)

```python
# Trying a modest range of C values (inverse regularization strength) to find optimal C for fitting a Logistic
C_list = [10**n for n in range(-5,5,1)]

# Creating empty list for training accuracy
train_acc =[]
# Creating empty list for validation accuracy
val_acc =[]

#Instantiating a scaler
scaler = StandardScaler()
#Fitting the scaler to the train set
scaler.fit(X_train)

#Scaling the training set
X_train_scaled = scaler.transform(X_train)
#Scaling the validation set
X_val_scaled = scaler.transform(X_val)

for value in C_list:
    
    #Instantiating a LogisticRegression model
    log_reg = LogisticRegression(C = value, random_state=3, n_jobs=-1, max_iter=10000 )
    
    #Fitting the model
    log_reg.fit(X_train_scaled, y_train)
        
    #Appending the trianing accuracy score to the list set up above
    train_acc.append(log_reg.score(X_train_scaled, y_train))
    val_acc.append(log_reg.score(X_val_scaled, y_val))
    
#Creating a df of the C values, training accuracy, & validation accuracy
train_val_df = pd.DataFrame({'C values': C_list,'training accuracy': train_acc , 'validation accuracy': val_acc})

#Showing the df of the C values, training accuracy, & validation accuracy
train_val_df
```

```python
#Plotting the accuracies for difference C values using a log scale for x-axis
fig = px.line(train_val_df, x='C values', y=['training accuracy','validation accuracy'])   
fig.update_layout(title = 'Train vs Validation Logistic Regression Accuracies for different C values', xaxis_title = 'C values', yaxis_title = 'Accuracy', xaxis_type="log")
fig.show()
```

```python
# -------------------------------------------------------
# Finding: C-Value Selection for Baseline Model
# -------------------------------------------------------

train_val_df['gap'] = abs(train_val_df['training accuracy'] - train_val_df['validation accuracy'])
best_row = train_val_df.loc[train_val_df['gap'].idxmin()]

# Identify the best C value (smallest gap between train and val accuracy)
best_C = best_row['C values']

print("**-------------------------------------------------------**")
print()
print("Finding")
print()
print(f"The C value with the smallest gap between training and validation accuracy")
print(f"is C = {best_C}, with:")
print(f"  - Training accuracy:   {best_row['training accuracy']:.4f}")
print(f"  - Validation accuracy: {best_row['validation accuracy']:.4f}")
print(f"  - Gap:                 {best_row['gap']:.4f}")
print()
print("This suggests good generalization with no strong evidence of overfitting.")
print("This C value will be used for the baseline logistic regression model below.")
print()
print("**-------------------------------------------------------**")
```

```python
#Instantiating a scaler
scaler = StandardScaler()
scaler.fit(X_rem)

X_rem_scaled = scaler.transform(X_rem)
X_test_scaled = scaler.transform(X_test)

#Instantiating a LogisticRegression model with the identified optimal C value
log_reg = LogisticRegression(C = best_C, random_state=3, n_jobs=-1, max_iter=10000 )
log_reg.fit(X_rem_scaled, y_rem)

rem_score = log_reg.score(X_rem_scaled, y_rem)
test_score = log_reg.score(X_test_scaled, y_test)

print(f"\nAccuracy score on train set:{round(rem_score,2)}\n\nAccuracy score on the test set:{round(test_score,2)}")
```

```python
# Getting predicted y values for the scaled test set
y_pred = log_reg.predict(X_test_scaled)

# Generating confusion matrix
cf_matrix = confusion_matrix(y_test, y_pred)
cf_df = pd.DataFrame(cf_matrix, columns=["Predicted Not Compatible", "Predicted Compatible"], index=["True Not Compatible", "True Compatible"])
display(cf_df)

ConfusionMatrixDisplay.from_estimator(log_reg, X_test_scaled, y_test)
plt.title("Compatibility Prediction with Baseline Model: Test Set Confusion Matrix")
plt.show()

prec_score = precision_score(y_test, y_pred)
rec_score = recall_score(y_test, y_pred)
f_score = f1_score(y_test, y_pred)

print(f"\nThe precision, recall, and F1 scores for the test set using the Baseline Model are:\n\n - Precision score: {prec_score}\n\n - Recall score: {rec_score}\n\n - F1 score: {f_score}")
```

```python
# -------------------------------------------------------
# Finding: Baseline Logistic Regression Performance
# -------------------------------------------------------

print("**-------------------------------------------------------**")
print()
print("Finding")
print()
print(f"Using this simple logistic regression model optimized for C = {best_C}:")
print()
print(f"  - The precision score of {prec_score:.2%} indicates that out of all the pairs")
print(f"    in the test dataset that the model classified as compatible,")
print(f"    {prec_score:.1%} of them are actually compatible.")
print()
print(f"  - The recall score of {rec_score:.2%} indicates that out of all the")
print(f"    actually compatible pairs in the test dataset, the model")
print(f"    accurately classified {rec_score:.1%} of them.")
print()
print(f"  - Given the class imbalance (43% compatible / 57% not compatible),")
print(f"    the F1 score of {f_score:.4f} indicates the overall balance between")
print(f"    precision and recall.")
print()
print("Insight")
print("  - For this dating compatibility model, we will prioritize")
print("    precision over recall. This means users will see")
print("    fewer incorrect or low-quality matches (false positives),")
print("    which helps build trust and satisfaction, even if it")
print("    means occasionally missing some good matches (false negatives).")
print()
print("**-------------------------------------------------------**")
```

```python
# Getting The baseline model probability for each point in the remainder and test sets
y_proba_train = log_reg.predict_proba(X_rem_scaled)[:, 1]
y_proba_test = log_reg.predict_proba(X_test_scaled)[:, 1]

fp_rate_train, tp_rate_train, thresholds_train = roc_curve(y_rem, y_proba_train)
roc_auc_train = roc_auc_score(y_rem, y_proba_train)
fp_rate_test, tp_rate_test, thresholds_test = roc_curve(y_test, y_proba_test)
roc_auc_test = roc_auc_score(y_test, y_proba_test)

plt.figure()
plt.plot(fp_rate_train, tp_rate_train, lw=2, label="Train set")
plt.plot(fp_rate_test, tp_rate_test, lw=2, label="Test set")
plt.plot([0, 1], [0, 1], lw=2, linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC And AUC for Baseline Model")
plt.legend(loc="best")
plt.show()

print(f"\nFor the Baseline Model:\n\n - AUC score for test set: {roc_auc_test:.4f}\n\n - AUC score for train set: {roc_auc_train:.4f}")
```

```python
# -------------------------------------------------------
# Finding: Baseline Model ROC and AUC Assessment
# -------------------------------------------------------

print("**-------------------------------------------------------**")
print()
print("Finding")
print()
print(f"Using this baseline simple logistic regression model optimized for C:")
print()
print(f"  - Both train and test ROC curves lie well above the diagonal random")
print(f"    line over almost the entire false positive rate range, so the model")
print(f"    is clearly better than random at distinguishing compatible vs not")
print(f"    compatible pairs.")
print()
print(f"  - The AUC score for the train set is {roc_auc_train:.4f} and for the")
print(f"    test set is {roc_auc_test:.4f}. These scores are close, suggesting")
print(f"    the model generalizes well and is not overfitting.")
print()

auc_gap = abs(roc_auc_train - roc_auc_test)
if auc_gap < 0.02:
    print(f"  - The gap between train and test AUC is only {auc_gap:.4f},")
    print(f"    confirming minimal overfitting.")
elif auc_gap < 0.05:
    print(f"  - The gap between train and test AUC is {auc_gap:.4f},")
    print(f"    which is acceptable but worth monitoring in subsequent models.")
else:
    print(f"  - The gap between train and test AUC is {auc_gap:.4f},")
    print(f"    which suggests some overfitting that should be addressed.")

print()
print("**-------------------------------------------------------**")
```

### <a id='toc7_2_'></a>7_2 Optimal Model Search and Hyperparameter Tuning: Improving Precision Score using Pipeline and RandomSearchCV [&#8593;](#toc0_)

```python
# -------------------------------------------------------
# Finding: Baseline Logistic Regression Performance
# -------------------------------------------------------

print("**-------------------------------------------------------**")
print()
print(f"  As mentioned earlier, Using this simple logistic regression model optimized for C = {best_C},")
print(f"  we achieve a solid precision of {prec_score:.2%}, a moderate recall of {rec_score:.1%}, and")
print(f"  an overall F1 score of (which sugegsts a reasonable balance between precision and recall)")
print()
print("**-------------------------------------------------------**")
```

We will now use a **Pipeline** combined with **RandomSearchCV** to:
- Systematically tune hyperparameters
- Compare different model types (Logistic Regression, Decision Tree, Random Forest, & KNN)
- Optimize for Precision score

```python
#Creating a base pipeline with preprocessing steps
estimators = [
    ("scaler", StandardScaler()),
    ("pca", PCA()),
    ("model", LogisticRegression())]

cachedir = mkdtemp()
model_pipe = Pipeline(estimators, memory=cachedir)

#Defining parameter grids for multiple model types
param_grid = [
    {
        "model": [LogisticRegression(random_state=3, max_iter=1000)],
        "scaler": [StandardScaler(), MinMaxScaler(), None],
        "pca__n_components": [0.7, 0.8, 0.9],
        "model__C": [best_C, 0.001, 0.01, 0.1, 1],
    },
    {
        "model": [DecisionTreeClassifier(random_state=3)],
        "scaler": [StandardScaler(), MinMaxScaler(), None],
        "pca__n_components": [0.6, 0.7, 0.8, 0.9],
        "model__max_depth": [5, 10, 15, 20],
    },
    {
        "model": [KNeighborsClassifier()],
        "scaler": [StandardScaler(), MinMaxScaler(), None],
        "pca__n_components": [0.7, 0.8, 0.9],
        "model__n_neighbors": [1, 2, 3, 4, 5],
    },
    {
        "model": [RandomForestClassifier(random_state=3)],
        "scaler": [StandardScaler(), MinMaxScaler(), None],
        "pca__n_components": [0.7, 0.8, 0.9],
        "model__n_estimators": list(range(10, 201, 10)),
        "model__max_depth": [5,10,15, None],
    },
    {
        "model": [GradientBoostingClassifier(random_state=3)],
        "scaler": [StandardScaler(), MinMaxScaler(), None],
        "pca__n_components": [0.7, 0.8, 0.9],
        "model__n_estimators": [50, 100, 150],
        "model__learning_rate": [0.01, 0.05, 0.1],
        "model__max_depth": [3, 5],
    }
]

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=3)
grid = RandomizedSearchCV(model_pipe, param_grid, cv=cv, scoring='precision', n_jobs=-1, verbose=3) #scoring is set to the performance feature to prioritize
fitted_grid = grid.fit(X_rem, y_rem)

print(f"\nBest parameters: {fitted_grid.best_params_}")
print(f"Best F1 score (cross-validated): {fitted_grid.best_score_:.4f}")

best_model = fitted_grid.best_estimator_
best_model
```

## <a id='toc8_'></a>8. Ideal Partner Range Generator [&#8593;](#toc0_)

Given a single user's profile, we score them against real partner profiles from the **holdout partner pool** (20% of the dataset reserved in Section 6_0), and extract the ranges of partner traits that consistently lead to `compatible = 1`.

This approach uses real profiles from the dataset, preserving natural feature correlations — trait combinations reflect realistic co-occurrences rather than independently sampled values.


### <a id='toc8_1_'></a>8_1 Scoring Against Holdout Partner Pool for a Given User Profile [&#8593;](#toc0_)


```python
# Define a sample user profile (Person A). This would come from the sign-up form in the MVP

user_profile = {
    'a_age': 29,
    'a_education': 3,
    'a_career_ambition': 0.70,
    'a_openness': 0.72,
    'a_extraversion': 0.55,
    'a_agreeableness': 0.68,
    'a_conscientiousness': 0.70,
    'a_chronotype': 0.80,
    'a_spontaneity': 0.35,
    'a_emotional_expressiveness': 0.55,
    'a_location': 'Urban',
    'a_career_field': 'Tech',
    'a_love_language': 'Quality Time',
    'a_gender': 'Female',
    'a_sexual_orientation': 'Straight'
}

print("User Profile (Person A):")
for key, val in user_profile.items():
    print(f"  {key}: {val}")
```

```python
# Extract compatible partners from the holdout pool using orientation-aware filtering
# This mitigates Risk 3 (logical pairing problem): only pair with valid partners

np.random.seed(42)

# --- Reconstruct categorical columns from OHE and drop the OHE columns ---
# The holdout pool has OHE columns (e.g., b_gender_Female, b_career_field_Tech)
# We reverse them back to categorical values, then drop the OHE columns
# so the partner pool has clean raw columns (matching the pre-encoding format)

def reverse_ohe(df, prefix):
    """Reconstruct a categorical column from its one-hot encoded columns, then drop the OHE columns."""
    ohe_cols = [col for col in df.columns if col.startswith(prefix)]
    subset = df[ohe_cols]
    has_value = subset.sum(axis=1) > 0
    result = pd.Series(np.nan, index=df.index, dtype='object')
    result[has_value] = subset.loc[has_value].idxmax(axis=1).str.replace(prefix, '', regex=False)
    # Create the categorical column name (e.g., 'b_gender_' -> 'b_gender')
    col_name = prefix.rstrip('_')
    df[col_name] = result
    # Drop the OHE columns
    df.drop(columns=ohe_cols, inplace=True)
    return df

reverse_ohe(holdout_partner_pool, 'b_gender_')
reverse_ohe(holdout_partner_pool, 'b_sexual_orientation_')
reverse_ohe(holdout_partner_pool, 'b_career_field_')
reverse_ohe(holdout_partner_pool, 'b_love_language_')
reverse_ohe(holdout_partner_pool, 'b_location_')

# Check for any NaN values from the reconstruction
for col in ['b_gender', 'b_sexual_orientation', 'b_career_field', 'b_love_language', 'b_location']:
    n_null = holdout_partner_pool[col].isna().sum()
    if n_null > 0:
        print(f"  WARNING: {col} has {n_null} null values after OHE reversal")

print("Reconstructed categorical columns from OHE:")
print(f"  Gender: {holdout_partner_pool['b_gender'].dropna().unique().tolist()}")
print(f"  Orientation: {holdout_partner_pool['b_sexual_orientation'].dropna().unique().tolist()}")
print(f"  Career field: {holdout_partner_pool['b_career_field'].dropna().unique().tolist()}")
print(f"  Love language: {holdout_partner_pool['b_love_language'].dropna().unique().tolist()}")
print(f"  Location: {holdout_partner_pool['b_location'].dropna().unique().tolist()}")
print(f"  Columns remaining: {len(holdout_partner_pool.columns)}")

# --- Determine valid partner genders based on user's orientation ---
user_orientation = user_profile['a_sexual_orientation']
user_gender = user_profile['a_gender']

if user_orientation == 'Straight':
    valid_partner_genders = ['Female'] if user_gender == 'Male' else ['Male']
elif user_orientation in ['Gay', 'Lesbian']:
    valid_partner_genders = [user_gender]
elif user_orientation in ['Bisexual', 'Pansexual', 'Queer', 'Questioning']:
    valid_partner_genders = ['Male', 'Female', 'Nonbinary', 'Other']
else:  # Asexual or other
    valid_partner_genders = ['Male', 'Female', 'Nonbinary', 'Other']

print(f"\nUser: {user_gender}, {user_orientation}")
print(f"Valid partner genders: {valid_partner_genders}")

# Determine valid partner orientations for each partner gender
def get_valid_partner_orientations(user_gender, user_orientation, partner_gender):
    """Return orientations that would make a partner open to this pairing."""
    if partner_gender != user_gender:  # opposite-sex pair
        return ['Straight', 'Bisexual', 'Pansexual', 'Queer', 'Questioning']
    else:  # same-sex pair
        if partner_gender == 'Male':
            return ['Gay', 'Bisexual', 'Pansexual', 'Queer', 'Questioning']
        elif partner_gender == 'Female':
            return ['Lesbian', 'Bisexual', 'Pansexual', 'Queer', 'Questioning']
        else:  # Nonbinary/Other
            return ['Bisexual', 'Pansexual', 'Queer', 'Questioning']

# Filter the holdout partner pool by gender and orientation compatibility
valid_partners = holdout_partner_pool[
    holdout_partner_pool['b_gender'].isin(valid_partner_genders)
].copy()

# Further filter by orientation compatibility
orientation_mask = pd.Series(False, index=valid_partners.index)
for pg in valid_partner_genders:
    valid_orients = get_valid_partner_orientations(user_gender, user_orientation, pg)
    mask = (valid_partners['b_gender'] == pg) & (valid_partners['b_sexual_orientation'].isin(valid_orients))
    orientation_mask = orientation_mask | mask

valid_partners = valid_partners[orientation_mask].reset_index(drop=True)

# Add the user profile as Person A columns (same for every row)
for key, val in user_profile.items():
    valid_partners[key] = val

n_pool_partners = len(valid_partners)

print(f"\nHoldout partner pool total: {len(holdout_partner_pool):,} profiles")
print(f"After orientation filtering: {n_pool_partners:,} valid partners")
print(f"\nPartner gender distribution:")
print(valid_partners['b_gender'].value_counts().to_string())
print(f"\nPartner orientation distribution:")
print(valid_partners['b_sexual_orientation'].value_counts().to_string())
display(valid_partners.sample(5))
```

```python
# Apply the same feature engineering pipeline as Section 4

# valid_partners now has clean raw categorical columns for both a_ and b_ sides
# OHE encode ALL categorical columns using the SAME encoder fitted in Section 4
pool_ohe_cols = [col for col in ohe_cols if col in valid_partners.columns]
ary_pool_encoded = one_hot_encoder.transform(valid_partners[pool_ohe_cols])
df_pool_encoded = pd.DataFrame(ary_pool_encoded, columns=encoded_feature_names, index=valid_partners.index)

# Drop raw categorical columns and add encoded ones
pool_ohe_df = pd.concat([valid_partners.drop(columns=pool_ohe_cols), df_pool_encoded], axis=1)

# Create delta features
for trait in ['openness', 'extraversion', 'agreeableness', 'conscientiousness']:
    pool_ohe_df[f'{trait}_delta'] = abs(pool_ohe_df[f'a_{trait}'] - pool_ohe_df[f'b_{trait}'])

pool_ohe_df['chronotype_delta'] = abs(pool_ohe_df['a_chronotype'] - pool_ohe_df['b_chronotype'])
pool_ohe_df['spontaneity_delta'] = abs(pool_ohe_df['a_spontaneity'] - pool_ohe_df['b_spontaneity'])
pool_ohe_df['ambition_delta'] = abs(pool_ohe_df['a_career_ambition'] - pool_ohe_df['b_career_ambition'])
pool_ohe_df['expressiveness_delta'] = abs(pool_ohe_df['a_emotional_expressiveness'] - pool_ohe_df['b_emotional_expressiveness'])
pool_ohe_df['age_gap'] = abs(pool_ohe_df['a_age'] - pool_ohe_df['b_age'])

# Create match flags using the raw categorical columns from valid_partners
pool_ohe_df['same_career'] = (valid_partners['a_career_field'] == valid_partners['b_career_field']).astype(int)
pool_ohe_df['same_love_language'] = (valid_partners['a_love_language'] == valid_partners['b_love_language']).astype(int)
pool_ohe_df['same_location'] = (valid_partners['a_location'] == valid_partners['b_location']).astype(int)
pool_ohe_df['same_education'] = (valid_partners['a_education'] == valid_partners['b_education']).astype(int)
pool_ohe_df['same_gender'] = (valid_partners['a_gender'] == valid_partners['b_gender']).astype(int)
pool_ohe_df['same_orientation'] = (valid_partners['a_sexual_orientation'] == valid_partners['b_sexual_orientation']).astype(int)

# Composite scores
pool_ohe_df['personality_similarity'] = pool_ohe_df[['openness_delta', 'extraversion_delta', 'agreeableness_delta', 'conscientiousness_delta']].mean(axis=1)
pool_ohe_df['lifestyle_similarity'] = pool_ohe_df[['chronotype_delta', 'spontaneity_delta']].mean(axis=1)

# Verify no duplicate columns
dupes = pool_ohe_df.columns[pool_ohe_df.columns.duplicated()].tolist()
if dupes:
    print(f"WARNING: Duplicate columns found: {dupes}")
else:
    print("No duplicate columns.")

print(f"Partner pool pairs after feature engineering: {pool_ohe_df.shape}")
```

```python
# Score all partner pool pairs using the trained best model

# Ensure columns match the training set (same order, same columns)
pool_features = pool_ohe_df[feature_columns]

# Get compatibility predictions and probabilities
pool_ohe_df['compatible_pred'] = best_model.predict(pool_features)
pool_ohe_df['compatible_proba'] = best_model.predict_proba(pool_features)[:, 1]

# Summary
n_compatible = pool_ohe_df['compatible_pred'].sum()
pct_compatible = n_compatible / n_pool_partners * 100

print(f"\nOf {n_pool_partners:,} holdout partner pool profiles:")
print(f"  {n_compatible:,} predicted compatible ({pct_compatible:.1f}%)")
print(f"  {n_pool_partners - n_compatible:,} predicted not compatible ({100 - pct_compatible:.1f}%)")
```

### <a id='toc8_2_'></a>8_2 Identifying Compatible Partner Ranges [&#8593;](#toc0_)

```python
# Extract the compatible partners and compute the 10th-90th percentile range for each numeric trait

compatible_partners = pool_ohe_df[pool_ohe_df['compatible_pred'] == 1]
all_partners = pool_ohe_df

# Define the partner B numeric features we want ranges for
partner_numeric_features = [
    'b_openness', 'b_extraversion', 'b_agreeableness', 'b_conscientiousness',
    'b_career_ambition', 'b_chronotype', 'b_spontaneity', 'b_emotional_expressiveness',
    'b_age', 'b_education'
]

# Compute ranges for compatible vs all partners
range_results = []

for feat in partner_numeric_features:
    range_results.append({
        'Feature': feat.replace('b_', ''),
        'Your Value': user_profile.get(feat.replace('b_', 'a_'), '—'),
        'Ideal Range Low (10th pctl)': compatible_partners[feat].quantile(0.10).round(2),
        'Ideal Range High (90th pctl)': compatible_partners[feat].quantile(0.90).round(2),
        'Compatible Avg': compatible_partners[feat].mean().round(2),  # Compatible Avg is the average trait value among partners the model predicted as compatible with the user
        'All Partners Avg': all_partners[feat].mean().round(2), # - All Partners Avg is the average trait value across all partners in the orientation-filtered pool (compatible and incompatible combined)
    })

range_df = pd.DataFrame(range_results)

print("\nIdeal Partner Ranges (10th-90th percentile of compatible partners):")
display(range_df)
```

```python
# Compute categorical preferences for compatible partners

catg_partner_features = ['b_career_field', 'b_love_language', 'b_location']

print("\nIdeal Partner Categorical Preferences (compatible vs all):\n")

catg_range_results = []

for feat in catg_partner_features:
    # Get distribution in compatible vs all
    compatible_dist = compatible_partners[feat] if feat in compatible_partners.columns else valid_partners.loc[compatible_partners.index, feat]
    all_dist = valid_partners[feat]
    
    # If the feature was OHE'd, reconstruct from the original valid_partners df
    if feat not in compatible_partners.columns:
        compatible_dist = valid_partners.loc[compatible_partners.index, feat]
    else:
        compatible_dist = compatible_partners[feat]
    
    comp_counts = compatible_dist.value_counts(normalize=True).round(3)
    all_counts = all_dist.value_counts(normalize=True).round(3)
    
    feature_name = feat.replace('b_', '')
    user_val = user_profile.get(feat.replace('b_', 'a_'), '—')
    top_match = comp_counts.index[0]
    
    print(f"  {feature_name} (yours: {user_val})")
    print(f"    Best match: {top_match} (makes up {comp_counts.iloc[0]*100:.1f}% of compatible partners)")
    if len(comp_counts) > 1:
        print(f"    Also good: {comp_counts.index[1]} (makes up {comp_counts.iloc[1]*100:.1f}% of compatible partners)")
    print()
```

```python
# Visualizing the ideal partner ranges as horizontal bar ranges with the user's value marked-OPTION2

n_numeric = len(range_df)
n_cols = 2
n_rows = (n_numeric + n_cols - 1) // n_cols

fig = make_subplots(rows=n_rows, cols=n_cols,
                    subplot_titles=range_df['Feature'].tolist(),
                    horizontal_spacing=0.15, vertical_spacing=0.12)

for i, (_, row) in enumerate(range_df.iterrows()):
    r = i // n_cols + 1
    c = i % n_cols + 1
    feat = row['Feature']
    low = row['Ideal Range Low (10th pctl)']
    high = row['Ideal Range High (90th pctl)']
    user_val = row['Your Value']

    # Ideal range bar
    fig.add_trace(go.Bar(
        y=[feat], x=[high - low], base=low, orientation='h',
        marker_color='lightblue', showlegend=False,
        hovertemplate=f'{feat}: {low} - {high}<extra></extra>'
    ), row=r, col=c)

    # User's value marker
    if pd.notna(user_val):
        fig.add_trace(go.Scatter(
            y=[feat], x=[float(user_val)], mode='markers',
            marker=dict(color='red', size=12, symbol='diamond'),
            showlegend=False,
            hovertemplate=f'Your {feat}: {user_val}<extra></extra>'
        ), row=r, col=c)

    # Compatible avg line
    fig.add_trace(go.Scatter(
        y=[feat], x=[row['Compatible Avg']], mode='markers',
        marker=dict(color='green', size=10, symbol='line-ns-open', line=dict(width=2)),
        showlegend=False,
        hovertemplate=f'Compatible avg: {row["Compatible Avg"]}<extra></extra>'
    ), row=r, col=c)

fig.update_layout(
    title='Ideal Partner Ranges by Feature (blue=range, red=you, green=compatible avg)',
    height=250 * n_rows, width=900
)

fig.update_yaxes(showticklabels=False)

# Map education numeric values to category labels
edu_labels = {1: 'High School', 2: 'Some College', 3: "Bachelor's", 4: "Master's", 5:
'Doctorate'}
for i, feat in enumerate(range_df['Feature']):
    if 'education' in feat.lower():
      r = i // n_cols + 1
      c = i % n_cols + 1
      fig.update_xaxes(
          tickvals=list(edu_labels.keys()),
          ticktext=list(edu_labels.values()),
          row=r, col=c
      )
      break

fig.show(renderer="iframe")
```

### <a id='toc8_4_'></a>8_4 Generating Plain-Language Ideal Partner Summary [&#8593;](#toc0_)

```python
# Generate a plain-language summary of the ideal partner profile
# This is what the end user would see in the product (translated from model output)

# Map feature names to user-friendly labels
friendly_labels = {
    'openness': 'Curiosity / Openness',
    'extraversion': 'Social Energy / Extraversion',
    'agreeableness': 'Empathy / Agreeableness',
    'conscientiousness': 'Diligent / Conscientiousness',
    'career_ambition': 'Career Ambition',
    'chronotype': 'Morning vs Night',
    'spontaneity': 'Planner vs Spontaneous',
    'emotional_expressiveness': 'Emotional Expressiveness',
    'age': 'Age',
    'education': 'Education Level'
}

# Rank features by how much the compatible range differs from the overall range
# (larger difference = more important for this user)
range_df['range_width_compatible'] = range_df['Ideal Range High (90th pctl)'] - range_df['Ideal Range Low (10th pctl)']
range_df['selectivity'] = 1 - range_df['range_width_compatible']  # narrower range = more selective = more important
range_df_sorted = range_df.sort_values('selectivity', ascending=False)

print("="*60)
print("YOUR IDEAL PARTNER PROFILE")
print("="*60)
print()

print("WHAT MATTERS MOST (ranked by selectivity):")
print("-"*60)

for rank, (_, row) in enumerate(range_df_sorted.iterrows(), 1):
    feat = row['Feature']
    label = friendly_labels.get(feat, feat)
    low = row['Ideal Range Low (10th pctl)']
    high = row['Ideal Range High (90th pctl)']
    user_val = row['Your Value']
    avg = row['Compatible Avg']
    
    # Determine if user's value is within the ideal range
    if user_val != '—':
        in_range = low <= float(user_val) <= high
        range_note = "(similar to you)" if in_range else "(different from you)"
    else:
        range_note = ""
    
    print(f"\n  {rank}. {label}")
    print(f"     Ideal range: {low} - {high} {range_note}")
    print(f"     Your value: {user_val} | Compatible partner avg: {avg}")

print("\n")
print("="*60)
```

## <a id='toc9_'></a>9. Conclusion [&#8593;](#toc0_)

**--------------------------------------------------------**

**Summary of Findings**

This project developed a compatibility prediction system with an "ideal partner range generator" for a dating app using the **modified** Cupid's Algorithm dataset of 100,000 pair records, which includes gender and sexual orientation features. The dataset was clean with zero missing values and zero duplicates. After one-hot encoding categorical features (including gender and sexual orientation) and engineering delta/similarity features (including `same_gender` and `same_orientation` match flags), exploratory analysis addressed three questions:

**EDA Q1: What are the strongest predictors of compatibility?** A combined ranking of all categorical (by gap in pp) and trait comparison (by correlation strength) features identified the single largest compatibility factor and produced a full predictor hierarchy.

**EDA Q2: Do the top predictors compound when both are aligned?** By isolating pairs where both the top categorical and top trait comparison predictors are aligned, we tested whether the combined compatibility rate is additive (sum of individual gains) or diminishing (less than the sum).

**EDA Q3: When the #1 predictor is unfavorable, which secondary factors best compensate?** Among pairs where the top predictor is misaligned, we identified which secondary trait comparison and categorical signals most differentiate compatible from incompatible pairs — informing how the model should weight features when the primary signal is absent.

**--------------------------------------------------------**

**Dataset Modification**

The original dataset was augmented with `a_gender`, `a_sexual_orientation`, `b_gender`, and `b_sexual_orientation` columns. Gender (Male, Female, Nonbinary, Other) and orientation (Straight, Gay, Lesbian, Bisexual, Pansexual, Asexual, Queer, Questioning) were assigned with logical pairing consistency and real-world proportions (Gallup 2023). See `Risks_Adding_Sexual_Orientation_Feature.md` for full risk assessment.

**--------------------------------------------------------**

**Model Performance**

- A baseline Logistic Regression model was used to establish benchmark metrics.
- Using Pipeline with RandomizedSearchCV, the best model was identified across multiple model types (Logistic Regression, Decision Tree, Random Forest, KNN, Gradient Boosting).
- Decision threshold tuning was performed to optimize the F1 score for the compatibility prediction use-case.
- The final model with optimized threshold provides the best balance between identifying compatible pairs and avoiding bad match recommendations.

**--------------------------------------------------------**

**Key Predictors**

The feature importance analysis reveals which personality traits, lifestyle preferences, and demographic factors most strongly predict compatibility. The engineered delta features (measuring similarity between partners) provide interpretable explanations for users. The `same_gender` and `same_orientation` match flags are expected to rank low, confirming they serve as pre-filters rather than compatibility predictors.

**--------------------------------------------------------**

**Ideal Partner Range Generator (Section 8)**

Section 8 takes a single user's profile (including gender and sexual orientation) and scores them against real partner profiles from a **20% holdout partner pool** — reserved from the dataset before model training to prevent data leakage. The holdout pool is filtered using **orientation-aware pairing logic** (only partners whose gender and orientation are compatible with the user's), then each user–partner pair is scored with the trained model. The 10th-90th percentile ranges of partner traits that lead to `compatible = 1` are extracted as an "ideal partner profile" with recommended ranges ranked by selectivity.

By using real profiles from the dataset rather than synthetically generated partners, this approach preserves natural feature correlations — trait combinations reflect realistic co-occurrences. The holdout pool provides a large candidate set (~20K-40K real profiles).

**--------------------------------------------------------**

**Ethical Considerations**

- The ideal partner ranges are statistical tendencies, not absolute rules — the system should communicate this clearly.
- Demographic features (age, location, education) should be carefully evaluated for potential bias in match recommendations.
- **Sexual orientation is used as a pre-filter for valid pairing, not as a compatibility predictor** — the model should not learn that any orientation is "more compatible" than another.
- The system provides data-backed insight to inform the user's own judgment, not to make decisions for them.
- Users deserve transparency about how the AI determines compatibility and how ranges are generated (holdout pool scoring, not prescriptive).

**--------------------------------------------------------**

**Recommendation**

For the dating app, the **Tuned Best Model with optimized decision threshold** is recommended for pair scoring, combined with the **ideal partner range generator** for user onboarding. This configuration provides the best balance between surfacing compatible pairs and avoiding bad match recommendations. The holdout partner pool approach ensures that ideal partner profiles are derived from real data profiles with natural feature correlations, and the orientation-aware filtering ensures realistically valid matches. This empowers users with data-backed insight into what kind of partner is likely to be compatible with them, so they can apply their own judgment rather than relying on opaque algorithms.
