# 1. Introduction

**Purpose**

The original Cupid Algorithm dataset (`cupid_algorithm_dataset_original.csv`) does not include gender or sexual orientation for Person A or Person B. The team identified that these features are needed for the SE/UX layer of the CupidMatch product — specifically to filter which profiles a user sees based on their orientation preferences.

This notebook adds four new columns to the dataset:
- `a_gender` and `b_gender` — Male, Female, Nonbinary, Other
- `a_sexual_orientation` and `b_sexual_orientation` — Straight, Gay, Lesbian, Bisexual, Pansexual, Asexual, Queer, Questioning

**Key Assumption**

The original dataset is synthetic — all features and compatibility scores were generated, not collected from real couples. Adding synthetic gender and orientation features therefore carries **low ground truth risk** since all features are on equal footing. The real concern is ensuring no artificial bias is introduced and that all pairings are logically valid.

**Recommendation**

Gender and sexual orientation should function as a **pre-filter** in the product ("only show me compatible matches within my orientation preference") rather than as a compatibility predictor in the ML model. This mirrors how real dating apps handle orientation — it's a gating filter, not a compatibility signal.

**Reference**

See `Risks_Adding_Sexual_Orientation_Feature.md` for the full risk assessment document.

# 4. Risk Assessment Summary

Before modifying the dataset, we assessed 5 risks. This section documents each risk and how it informs our implementation approach.

<a id='section_4_1'></a>
### 4_1. Risk 1: No Ground Truth

The original dataset was generated *without* gender or orientation influencing the compatibility scores. Adding these synthetically means they have **zero real relationship to the target variable** (`compatible`).

**Assessment:** LOW risk. The original dataset is itself synthetic — all features are generated. Since every feature is on equal footing, this risk is negligible. The model will likely ignore these features (low importance), which is acceptable since they serve as UX pre-filters, not compatibility predictors.

<a id='section_4_2'></a>
### 4_2. Risk 2: Bias Amplification

If gender/orientation are assigned with non-uniform distributions (mirroring real demographics), the model could learn that certain orientations or genders are "more compatible" simply because they appear more frequently — a **class imbalance bias** problem.

**Mitigation strategy:**
- Use real-world proportions (Gallup 2023) but slightly overrepresent minority categories to reduce extreme imbalance
- **Validate** that compatibility rates are uniform across all categories after assignment
- Accept small deviations due to random variation, but flag anything beyond ~3 percentage points

<a id='section_4_3'></a>
### 4_3. Risk 3: Logical Pairing Problem

Sexual orientation determines *who should be paired in the first place*, not compatibility *within* a pair. Random assignment would create impossible pairings (e.g., a Straight man paired with another man).

**Mitigation strategy:**
- Assign pair type (gender combo) **first**
- Then assign orientations that are **consistent with that pairing**
- Enforce 5 logical rules:
  1. Straight only in opposite-sex binary pairs
  2. Gay only for males in M-M pairs
  3. Lesbian only for females in F-F pairs
  4. Nonbinary/Other persons are never assigned Straight
  5. Binary persons paired with NB/Other cannot be Straight, Gay, or Lesbian
- Bisexual, Pansexual, Queer, Questioning, and Asexual are **universal** — valid in any pair type

<a id='section_4_4'></a>
### 4_4. Risk 4: Ethical / Presentation Risk

A model that treats sexual orientation as a *compatibility predictor* could be interpreted as saying some orientations are more or less compatible. This is a sensitive area.

**Mitigation strategy:**
- Use orientation as a **pre-filter** in the product layer, not as a model feature
- Validate that the model does not surface orientation-based compatibility differences
- In sprint demos, clearly communicate that orientation filtering happens *before* the model scores compatibility

<a id='section_4_5'></a>
### 4_5. Risk 5: Feature Leakage Risk

If orientation is assigned and then pairs are filtered to only "valid" combos, a selection bias could be introduced that the model exploits.

**Mitigation strategy:**
- We do NOT filter any pairs — all 100,000 rows are preserved
- Instead, we assign orientations that are *consistent with* the existing pair, avoiding the need to remove any data

# 5. Feature Creation: Gender Assignment

<a id='section_5_1'></a>
### 5_1. Define Pair Type Distribution

Gender must be assigned before orientation because orientation logic depends on the gender combination of each pair.

**Proportions rationale (based on Gallup 2023 data):**
- ~93% of US couples are opposite-sex → we use 84% (slightly reduced to give more representation to other pair types)
- ~7% of US couples are same-sex → we use 9% (split evenly M-M and F-F)
- ~1.6% of US adults identify as transgender/nonbinary → we use ~5% for NB-involved pairs and ~2% for Other-involved pairs

These proportions are intentionally adjusted from exact real-world values to **reduce extreme class imbalance** (Risk 2 mitigation) while remaining realistic.

```python
#Setting random seed for reproducibility
np.random.seed(42)

n = len(cupid_df)

#Defining pair type probabilities
#Slightly overrepresenting minority pair types to mitigate Risk 2 (bias amplification)
pair_type_probs = {
    'opposite': 0.84,      # M-F or F-M pairs
    'mm': 0.045,           # Male-Male pairs
    'ff': 0.045,           # Female-Female pairs
    'nb_involved': 0.03,   # Nonbinary paired with Male, Female, or Other
    'other_involved': 0.02,# Other gender paired with Male, Female, or NB
    'nb_nb': 0.02,         # Nonbinary-Nonbinary pairs
}

print("Pair type probability distribution:")
for pair_type, prob in pair_type_probs.items():
    print(f"  {pair_type:20s}: {prob*100:.1f}%")
print(f"  {'TOTAL':20s}: {sum(pair_type_probs.values())*100:.1f}%")

#Assigning pair types to each row
pair_types = np.random.choice(
    list(pair_type_probs.keys()),
    size=n,
    p=list(pair_type_probs.values())
)

print(f"\nPair types assigned to {n} rows")
```

<a id='section_5_2'></a>
### 5_2. Assign Genders Based on Pair Type

Each pair type maps to specific gender combinations. For opposite-sex pairs, we randomly assign which person is Male vs Female to avoid always making Person A the same gender.

```python
#Assigning genders based on pair type
a_gender = []
b_gender = []

for pt in pair_types:
    if pt == 'opposite':
        # Randomly assign which is male/female (avoids always making A=male)
        if np.random.random() < 0.5:
            a_gender.append('Male')
            b_gender.append('Female')
        else:
            a_gender.append('Female')
            b_gender.append('Male')
    elif pt == 'mm':
        a_gender.append('Male')
        b_gender.append('Male')
    elif pt == 'ff':
        a_gender.append('Female')
        b_gender.append('Female')
    elif pt == 'nb_involved':
        # One person is Nonbinary, the other is Male or Female
        if np.random.random() < 0.5:
            a_gender.append('Nonbinary')
            b_gender.append(np.random.choice(['Male', 'Female']))
        else:
            a_gender.append(np.random.choice(['Male', 'Female']))
            b_gender.append('Nonbinary')
    elif pt == 'other_involved':
        # One person identifies as Other
        if np.random.random() < 0.5:
            a_gender.append('Other')
            b_gender.append(np.random.choice(['Male', 'Female', 'Nonbinary']))
        else:
            a_gender.append(np.random.choice(['Male', 'Female', 'Nonbinary']))
            b_gender.append('Other')
    elif pt == 'nb_nb':
        a_gender.append('Nonbinary')
        b_gender.append('Nonbinary')

cupid_df['a_gender'] = a_gender
cupid_df['b_gender'] = b_gender

print("Gender columns assigned successfully")
print(f"Shape: {cupid_df.shape}")
```

# 6. Feature Creation: Sexual Orientation Assignment

<a id='section_6_1'></a>
### 6_1. Define Pairing Rules

Orientation must be logically consistent with the gender pairing (Risk 3 mitigation). The rules are:

| Rule | Description |
|---|---|
| **Straight** | Only valid for Male or Female persons in opposite-sex (M-F / F-M) pairs |
| **Gay** | Only valid for Male persons in M-M pairs |
| **Lesbian** | Only valid for Female persons in F-F pairs |
| **Bisexual** | Valid in any pair type |
| **Pansexual** | Valid in any pair type |
| **Asexual** | Valid in any pair type (can still have romantic relationships) |
| **Queer** | Valid in any pair type (umbrella term) |
| **Questioning** | Valid in any pair type (exploring) |
| **NB/Other persons** | Never assigned Straight; can be any other orientation |
| **Binary person + NB/Other partner** | Cannot be Straight, Gay, or Lesbian |

```python
#Defining the function that returns valid orientations based on the gender pairing
#This is the core Risk 3 mitigation logic

def get_valid_orientations(person_gender, partner_gender):
    """Return valid orientations for a person given their gender and partner's gender.
    
    Rules:
    - Bisexual, Pansexual, Queer, Questioning, Asexual are UNIVERSAL (valid in any pair)
    - Straight only for Male/Female in opposite-sex binary pairs
    - Gay only for Males in M-M pairs
    - Lesbian only for Females in F-F pairs
    - NB/Other persons: any orientation except Straight
    - Binary person dating NB/Other: only universal orientations
    """
    
    # Universal orientations valid in any pairing
    universal = ['Bisexual', 'Pansexual', 'Queer', 'Questioning', 'Asexual']
    
    # Determine pairing type
    is_same_binary = (person_gender == partner_gender and person_gender in ['Male', 'Female'])
    is_opposite_binary = (person_gender != partner_gender and
                          person_gender in ['Male', 'Female'] and
                          partner_gender in ['Male', 'Female'])
    involves_nb_other = (person_gender in ['Nonbinary', 'Other'] or
                         partner_gender in ['Nonbinary', 'Other'])
    
    if involves_nb_other:
        if person_gender in ['Nonbinary', 'Other']:
            # NB/Other can identify as any orientation except Straight
            return universal + ['Gay', 'Lesbian']
        else:
            # Binary person dating NB/Other: only universal orientations
            return universal
    elif is_opposite_binary:
        return ['Straight'] + universal
    elif is_same_binary:
        if person_gender == 'Male':
            return ['Gay'] + universal
        else:
            return ['Lesbian'] + universal
    else:
        return universal

print("Pairing rules function defined")
print("\nExample outputs:")
test_cases = [
    ('Male', 'Female', 'Opposite-sex'),
    ('Female', 'Male', 'Opposite-sex'),
    ('Male', 'Male', 'Same-sex male'),
    ('Female', 'Female', 'Same-sex female'),
    ('Nonbinary', 'Male', 'NB + binary'),
    ('Male', 'Nonbinary', 'Binary + NB'),
    ('Other', 'Female', 'Other + binary'),
    ('Nonbinary', 'Nonbinary', 'NB + NB'),
]
for pg, partg, label in test_cases:
    valid = get_valid_orientations(pg, partg)
    print(f"  {label:20s} ({pg} + {partg}): {valid}")
```

```python
#Defining the probability-weighted assignment function
#Base weights approximate real-world proportions (Gallup 2023)
#Minority orientations are slightly overrepresented to mitigate Risk 2

def assign_orientation_with_weights(valid_orientations):
    """Assign orientation with realistic probability weights from the valid options."""
    
    # Base weights for each orientation (approximate real-world proportions)
    base_weights = {
        'Straight': 80,
        'Bisexual': 8,
        'Gay': 3,
        'Lesbian': 2.5,
        'Pansexual': 2.5,
        'Queer': 1.5,
        'Asexual': 1.5,
        'Questioning': 1.0,
    }
    
    # Get weights only for valid orientations, then normalize to sum to 1
    weights = [base_weights[o] for o in valid_orientations]
    total = sum(weights)
    probs = [w / total for w in weights]
    
    return np.random.choice(valid_orientations, p=probs)

print("Probability-weighted assignment function defined")
print("\nBase weights:")
base_weights = {'Straight': 80, 'Bisexual': 8, 'Gay': 3, 'Lesbian': 2.5, 'Pansexual': 2.5, 'Queer': 1.5, 'Asexual': 1.5, 'Questioning': 1.0}
for orient, weight in base_weights.items():
    print(f"  {orient:15s}: {weight}")
```

<a id='section_6_2'></a>
### 6_2. Assign Orientations Consistent with Pairing

```python
#Assigning sexual orientation for each person, consistent with their gender pairing
#This is the core Risk 3 mitigation: orientation is NEVER randomly assigned
#It is always constrained to be logically valid given the pair's gender combination

a_orientation = []
b_orientation = []

for i in range(n):
    ag = cupid_df.iloc[i]['a_gender']
    bg = cupid_df.iloc[i]['b_gender']
    
    # Get valid orientations for each person given their partner's gender
    valid_a = get_valid_orientations(ag, bg)
    valid_b = get_valid_orientations(bg, ag)
    
    # Assign with probability weights
    a_orientation.append(assign_orientation_with_weights(valid_a))
    b_orientation.append(assign_orientation_with_weights(valid_b))

cupid_df['a_sexual_orientation'] = a_orientation
cupid_df['b_sexual_orientation'] = b_orientation

print(f"Sexual orientation assigned to {n} pairs")
print(f"Shape: {cupid_df.shape}")
```

# 7. Risk 3 Validation: Logical Pairing Checks

This is the critical validation step. We run 5 checks to confirm that no logically impossible pairings exist. **All counts should be 0.**

```python
#--------------------------------------------------------------
# CHECK 1: No Straight person in a same-sex binary pair
#--------------------------------------------------------------
same_sex_binary = cupid_df[
    ((cupid_df['a_gender'] == 'Male') & (cupid_df['b_gender'] == 'Male')) |
    ((cupid_df['a_gender'] == 'Female') & (cupid_df['b_gender'] == 'Female'))
]
straight_in_same_sex = same_sex_binary[
    (same_sex_binary['a_sexual_orientation'] == 'Straight') |
    (same_sex_binary['b_sexual_orientation'] == 'Straight')
]
print(f"CHECK 1: Straight person in same-sex binary pair")
print(f"  Same-sex binary pairs: {len(same_sex_binary):,}")
print(f"  Violations: {len(straight_in_same_sex)} {'PASS' if len(straight_in_same_sex) == 0 else 'FAIL'}")
```

```python
#--------------------------------------------------------------
# CHECK 2: No Gay person in an opposite-sex pair
#--------------------------------------------------------------
opp_sex = cupid_df[
    ((cupid_df['a_gender'] == 'Male') & (cupid_df['b_gender'] == 'Female')) |
    ((cupid_df['a_gender'] == 'Female') & (cupid_df['b_gender'] == 'Male'))
]
gay_in_opp = opp_sex[
    (opp_sex['a_sexual_orientation'] == 'Gay') |
    (opp_sex['b_sexual_orientation'] == 'Gay')
]
print(f"CHECK 2: Gay person in opposite-sex pair")
print(f"  Opposite-sex pairs: {len(opp_sex):,}")
print(f"  Violations: {len(gay_in_opp)} {'PASS' if len(gay_in_opp) == 0 else 'FAIL'}")
```

```python
#--------------------------------------------------------------
# CHECK 3: No Lesbian person in an opposite-sex pair
#--------------------------------------------------------------
lesbian_in_opp = opp_sex[
    (opp_sex['a_sexual_orientation'] == 'Lesbian') |
    (opp_sex['b_sexual_orientation'] == 'Lesbian')
]
print(f"CHECK 3: Lesbian person in opposite-sex pair")
print(f"  Opposite-sex pairs: {len(opp_sex):,}")
print(f"  Violations: {len(lesbian_in_opp)} {'PASS' if len(lesbian_in_opp) == 0 else 'FAIL'}")
```

```python
#--------------------------------------------------------------
# CHECK 4: No Nonbinary/Other person identifying as Straight
#--------------------------------------------------------------
nb_other_straight = cupid_df[
    ((cupid_df['a_gender'].isin(['Nonbinary', 'Other'])) & (cupid_df['a_sexual_orientation'] == 'Straight')) |
    ((cupid_df['b_gender'].isin(['Nonbinary', 'Other'])) & (cupid_df['b_sexual_orientation'] == 'Straight'))
]
print(f"CHECK 4: Nonbinary/Other person identifying as Straight")
print(f"  Violations: {len(nb_other_straight)} {'PASS' if len(nb_other_straight) == 0 else 'FAIL'}")
```

```python
#--------------------------------------------------------------
# CHECK 5: No Straight binary person paired with NB/Other
#--------------------------------------------------------------
nb_other_pairs = cupid_df[
    (cupid_df['a_gender'].isin(['Nonbinary', 'Other'])) | (cupid_df['b_gender'].isin(['Nonbinary', 'Other']))
]
straight_with_nb = nb_other_pairs[
    ((~nb_other_pairs['a_gender'].isin(['Nonbinary', 'Other'])) & (nb_other_pairs['a_sexual_orientation'] == 'Straight')) |
    ((~nb_other_pairs['b_gender'].isin(['Nonbinary', 'Other'])) & (nb_other_pairs['b_sexual_orientation'] == 'Straight'))
]
print(f"CHECK 5: Straight binary person paired with NB/Other")
print(f"  Pairs involving NB/Other: {len(nb_other_pairs):,}")
print(f"  Violations: {len(straight_with_nb)} {'PASS' if len(straight_with_nb) == 0 else 'FAIL'}")
```

```python
#--------------------------------------------------------------
# SUMMARY: All 5 checks
#--------------------------------------------------------------
checks = [
    ('Straight in same-sex binary pair', len(straight_in_same_sex)),
    ('Gay in opposite-sex pair', len(gay_in_opp)),
    ('Lesbian in opposite-sex pair', len(lesbian_in_opp)),
    ('NB/Other person as Straight', len(nb_other_straight)),
    ('Straight binary + NB/Other partner', len(straight_with_nb)),
]

print("="*60)
print("RISK 3 VALIDATION SUMMARY")
print("="*60)
all_pass = True
for check_name, violations in checks:
    status = 'PASS' if violations == 0 else 'FAIL'
    if violations > 0:
        all_pass = False
    print(f"  {check_name:45s}: {violations} violations [{status}]")
print("="*60)
print(f"Overall: {'ALL CHECKS PASSED' if all_pass else 'SOME CHECKS FAILED'}")
print(f"Total rows preserved: {len(cupid_df):,} (no rows lost)")
```

<a id='section_8'></a>
# 8. Risk 2 Validation: Bias Amplification Checks

We verify that compatibility rates are approximately uniform across all gender and orientation categories. If any category has a substantially different compatibility rate, it would indicate artificial bias introduced by the assignment process.

```python
#Checking compatibility rate by sexual orientation (Person A)
print("--- Compatibility Rate by Sexual Orientation (Person A) ---\n")

orient_compat = []
for orient in sorted(cupid_df['a_sexual_orientation'].unique()):
    subset = cupid_df[cupid_df['a_sexual_orientation'] == orient]
    rate = subset['compatible'].mean() * 100
    orient_compat.append({'Orientation': orient, 'Compatibility Rate (%)': round(rate, 1), 'Sample Size': len(subset)})
    print(f"  {orient:15s}: {rate:.1f}% compatible (n={len(subset):,})")

orient_compat_df = pd.DataFrame(orient_compat)
rate_range = orient_compat_df['Compatibility Rate (%)'].max() - orient_compat_df['Compatibility Rate (%)'].min()
print(f"\n  Range: {rate_range:.1f} percentage points")
print(f"  Assessment: {'ACCEPTABLE (<5 pp)' if rate_range < 5 else 'CONCERNING (>=5 pp) - investigate further'}")
```

```python
#Visualizing compatibility rate by orientation
fig = px.bar(orient_compat_df, x='Orientation', y='Compatibility Rate (%)',
             title='Compatibility Rate by Sexual Orientation (Risk 2 Check)',
             color='Compatibility Rate (%)',
             color_continuous_scale='RdYlGn',
             text='Compatibility Rate (%)')
fig.update_traces(textposition='outside')
fig.update_layout(yaxis_range=[0, 60])
fig.show()
```

```python
#Checking compatibility rate by gender (Person A)
print("--- Compatibility Rate by Gender (Person A) ---\n")

gender_compat = []
for gender in sorted(cupid_df['a_gender'].unique()):
    subset = cupid_df[cupid_df['a_gender'] == gender]
    rate = subset['compatible'].mean() * 100
    gender_compat.append({'Gender': gender, 'Compatibility Rate (%)': round(rate, 1), 'Sample Size': len(subset)})
    print(f"  {gender:12s}: {rate:.1f}% compatible (n={len(subset):,})")

gender_compat_df = pd.DataFrame(gender_compat)
rate_range_g = gender_compat_df['Compatibility Rate (%)'].max() - gender_compat_df['Compatibility Rate (%)'].min()
print(f"\n  Range: {rate_range_g:.1f} percentage points")
print(f"  Assessment: {'ACCEPTABLE (<5 pp)' if rate_range_g < 5 else 'CONCERNING (>=5 pp) - investigate further'}")
```

```python
#Visualizing compatibility rate by gender
fig = px.bar(gender_compat_df, x='Gender', y='Compatibility Rate (%)',
             title='Compatibility Rate by Gender (Risk 2 Check)',
             color='Compatibility Rate (%)',
             color_continuous_scale='RdYlGn',
             text='Compatibility Rate (%)')
fig.update_traces(textposition='outside')
fig.update_layout(yaxis_range=[0, 60])
fig.show()
```

# 9. Column Ordering and Final Inspection

Reorder columns so that `a_gender` and `a_sexual_orientation` are placed right after Person A's existing features, and `b_gender` and `b_sexual_orientation` after Person B's features. Target variables remain at the end.

```python
#Reordering columns: new gender/orientation columns placed after each person's feature block
a_original = [c for c in cupid_df.columns if c.startswith('a_') and c not in ['a_gender', 'a_sexual_orientation']]
b_original = [c for c in cupid_df.columns if c.startswith('b_') and c not in ['b_gender', 'b_sexual_orientation']]

col_order = (
    ['pair_id']
    + a_original
    + ['a_gender', 'a_sexual_orientation']
    + b_original
    + ['b_gender', 'b_sexual_orientation']
    + ['compatibility_score', 'compatible', 'relationship_longevity_months']
)

cupid_df = cupid_df[col_order]

print(f"Final column order ({len(cupid_df.columns)} columns):")
for i, col in enumerate(cupid_df.columns):
    new_marker = ' [NEW]' if col in ['a_gender', 'a_sexual_orientation', 'b_gender', 'b_sexual_orientation'] else ''
    print(f"  {i+1:2d}. {col}{new_marker}")
```

```python
#Final inspection of the modified dataset
print(f"Shape: {cupid_df.shape}")
print(f"Missing values: {cupid_df.isnull().sum().sum()}")
print(f"Duplicated rows: {cupid_df.duplicated().sum()}")
print(f"\nSample rows:")
display(cupid_df.sample(10))
```
